```python
# bfs
class Solution:
    def solveNQueens(self, n: int) -> [[str]]:
        q = []
        for i in range(n):
            q += [[i]]

        res = []
        while q:
            c = q.pop()
            l = len(c)
            if l == n:
                res += [c]
                continue
            for i in range(n):
                if all(i != j and l - k != abs(j - i) for k, j in enumerate(c)):
                    q.insert(0, c + [i])

        op = []
        s = '.' * n
        for r in res:
            opc = []
            for e in r:
                opc += [s[:e] + 'Q' + s[e+1:]]
            op += [opc]

        return op
```

```python
sol = Solution()
```

```python
sol.solveNQueens(8)
```

```python
#UCS

def isSafe (board, row, col):
    # check left row
    for y in range(col):
        if board[row][y] == 1:
            return False
    # check diagonal left top
    for x, y in zip(range(row, -1, -1), range(col, -1, -1)):
        if board[x][y] == 1:
            return False
    # check diagonal left bottom
    for x, y in zip(range(row, N, 1), range(col, -1, -1)):
        if board[x][y] == 1:
            return False

    return True

def generateSolution(board, col):
    # terminating condition
    # all columns covered
    global N
    if col >= N:
        return True
    # loop over all the rows
    for i in range(N):
        if isSafe(board, i, col) == True:
            board[i][col] = 1
            # recursively place other queens
            if generateSolution(board, col + 1) == True:
                return True
            # unmark queen spot
            board[i][col] = 0
    # backtrack
    return False


N = 8
startCol = 0
board = [[0 for i in range(N)] for j in range(N)]
# print(board)

if generateSolution(board, startCol) == False:
    print("No Solution Exists")
else:
    print("Solution exists")
    print(board)
```
